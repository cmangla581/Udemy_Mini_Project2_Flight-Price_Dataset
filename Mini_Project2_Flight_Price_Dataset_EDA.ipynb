```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns 
```

 Now, we will be performing the feature engineering and the EDA simultaneously as: 

```python
df['Date'] = df['Date_of_Journey'].str.split("/").str[0]  
df['Date'] 

# Here all the dates have been received seperately 
```

```python
# Now, suppose we want the month seperately, then it can be doen as: 
df['Month'] = df['Date_of_Journey'].str.split("/").str[1]
df['Month'] 
```

```python
# Now, if we want the year seperately, then it occurs as: 
df['Year'] = df['Date_of_Journey'].str.split("/").str[2] 
df['Year']  

# All the years have been captures seperately 
```

```python
# Now, if we print the data 
df.head()  

# Here, as we can see that 3 extra columns of date, month and year have been successfully addded in the table 
```

```python
df.info()  

# Here, as we can see that  the date month and year columns are still as the object datatypes 
```

```python
# Now, foe successfully converting them in the int datatype, we can write the code as: 
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()  

# Here it has been converted to iteger data type 
```

```python
# Now, if we have to drop the date of journey column, then we carry it out as: 
df.drop('Date_of_Journey', axis = 1, inplace = True)  
```

```python
df.head()  

# here, as we can see that the date of journey has been removed properly from the table 
```

```python
# First we seperate only the time from the function as: 

df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x:x.split(' ')[0]) 
```

Like, we did with the date, month and the year column, similarly, we try this with the the hour and the minute column to seperate them make seperate columns for them. 

```python
df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0] 
df["Arrival_min"] = df["Arrival_Time"].str.split(":").str[1]  
```

```python
# Now, presenting the whole dataset as: 
df.head()  

# Here, seperate column of hour and minutes has been made  
# Also as we can see that the hour and the minute has been seperated 
```

```python
df.info()  

# Here, as we can see that the hour and the min are seperated but the min and hour are of object type 
```

```python
df["Arrival_hour"] = df["Arrival_hour"].astype(int)
df["Arrival_min"] = df["Arrival_min"].astype(int) 
```

```python
df.info()  

# Here, the arrival time and arrival min has been converted to int type 
```

```python
# Now, if we have to drop the arrivaal time, then we go on as:  

df.drop('Arrival_Time', axis = 1, inplace = True) 
```

```python
df.head()  

# Here, as we can see that the arrival time column has been removed from the table 
```

Like, we did for the time and the hour, we can do the same for the departure time 

```python
df["Departure_hour"] = df["Dep_Time"].str.split(":").str[0] 
df["Departure_min"] = df["Dep_Time"].str.split(":").str[1] 
```

```python
df.head() 
# Here the head of the final data will be presented here 
```

```python
# Now, cnverting them into the integer data type follows as: 
df["Departure_hour"] = df["Departure_hour"].astype(int)
df["Departure_min"] = df["Departure_min"].astype(int)
```

```python
df.info()  

# Here, both the departure time of hour and minutes has been converted to int data type. 
```

```python
# If we have to drop the dep_time column, then it goes on as: 
df.drop('Dep_Time', axis = 1, inplace = True) 
```

```python
df.head()  
# Dep Time column has been removed from here 
```

```python
# If we want the unique values of the total stops column, then it goes on as: 
df['Total_Stops'].unique()  

# Here all the unique values get printed 
```

```python
df.isnull().sum()  

# Here as we can see the total null values present 
```

```python
# Now, if we want the null value from the  total_stops column, then it goes on as: 
df[df['Total_Stops'].isnull()]  

# Here we have got the NAN value present here 
```

Now, we will be doing the mapping so, as to convert the stops data into the numericals, which will allow for the easy data processing as:  

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan:1}) 
```

```python
df[df['Total_Stops'].isnull()]  

# Here all the data points are non filled after the proper execution 
```

```python
df.head()  
# Here, the total stops has the specific label as follows 
```

```python
# Now,if we have to drop the route column, then it goes as: 
df.drop('Route', axis = 1, inplace = True) 
```

Here, we can cconvert all the categorical data, into the numerical by using the One Hot Encoder as: 


```python
from sklearn.preprocessing import OneHotEncoder 

encoder = OneHotEncoder() 
```

```python
encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()  

# Here it has been converted to the array 
```

Now, converting it into the data Frame goes on as:  

```python
df2 = pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), columns = encoder.get_feature_names_out())
# Here all the data has been converted properly here 
```
